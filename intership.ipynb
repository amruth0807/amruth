```python
source_counts = df['Source'].value_counts()
print("Information Sources and their Counts:\n", source_counts)

print("\nSummary of Common Sources:")
if not source_counts.empty:
    most_common_source = source_counts.idxmax()
    most_common_count = source_counts.max()
    print(f"The most common source of investment information is '{most_common_source}' with {most_common_count} mentions.")
    if len(source_counts) > 1:
        second_most_common_source = source_counts.drop(most_common_source).idxmax()
        second_most_common_count = source_counts.drop(most_common_source).max()
        print(f"The second most common source is '{second_most_common_source}' with {second_most_common_count} mentions.")
else:
    print("No information sources found in the dataset.")
```

# Task
Analyze the correlations between 'age', 'Duration_Numerical', and a numerical representation of the 'Expect' column in the `df` DataFrame by converting the 'Expect' column from categorical ranges to numerical midpoints, calculating the Pearson correlation matrix for these three columns, and then visualizing the matrix using a heatmap.

## Convert 'Expect' column to numerical format

### Subtask:
Map the categorical ranges in the 'Expect' column (e.g., '20%-30%') to numerical midpoints or representative values for correlation analysis.


**Reasoning**:
I will create a dictionary to map the categorical 'Expect' values to their numerical midpoints and then apply this mapping to create a new 'Expect_Numerical' column in the DataFrame.



```python
expect_mapping = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35
}

df['Expect_Numerical'] = df['Expect'].map(expect_mapping)

print(df[['Expect', 'Expect_Numerical']].head())
```

**Reasoning**:
Now that the 'Expect' column has been converted to numerical format as 'Expect_Numerical', the next step is to calculate the Pearson correlation matrix between 'age', 'Duration_Numerical', and 'Expect_Numerical'.



```python
correlation_matrix = df[['age', 'Duration_Numerical', 'Expect_Numerical']].corr(method='pearson')
print("Pearson Correlation Matrix:")
print(correlation_matrix)
```

**Reasoning**:
To visualize the relationships between the selected numerical features, I will create a heatmap of the Pearson correlation matrix using `seaborn` and `matplotlib.pyplot`.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Age, Duration, and Expectations')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The categorical 'Expect' column, with ranges like '10%-20%', '20%-30%', and '30%-40%', was successfully converted into a numerical representation using midpoints (15, 25, and 35, respectively) and stored in a new column named `Expect_Numerical`.
*   The Pearson correlation analysis revealed a moderate positive correlation between `Duration_Numerical` and `Expect_Numerical` (0.258). This suggests that as the duration increases, the expected percentage also tends to increase.
*   'age' showed a very weak positive correlation with `Duration_Numerical` (0.052) and a weak negative correlation with `Expect_Numerical` (-0.090), indicating minimal linear relationships between age and these other variables.
